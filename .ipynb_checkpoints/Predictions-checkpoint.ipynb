```python
import pandas as pd
import numpy as np
from fbprophet import Prophet
```

```python
walmart = pd.read_csv('walmartdata.csv')
walmart.head()
```

```python
fuel = []
unem = []
temp = []
cpi = []
for i in range((walmart["Store"].max()) + 1):
  walmar = walmart.loc[walmart['Store'] == i]
  df = walmar.filter(items=['Start_of_Week', 'Fuel_Price'])
  df = df.rename(columns={"Start_of_Week": "ds", "Fuel_Price": "y"})
  df = df.reset_index()
  df = df.drop(['index'], axis=1)
  fuel.append(df)
  df = walmar.filter(items=['Start_of_Week', 'Unemployment'])
  df = df.rename(columns={"Start_of_Week": "ds", "Unemployment": "y"})
  df = df.reset_index()
  df = df.drop(['index'], axis=1)
  unem.append(df)
  df = walmar.filter(items=['Start_of_Week', 'Temperature_C'])
  df = df.rename(columns={"Start_of_Week": "ds", "Temperature_C": "y"})
  df = df.reset_index()
  df = df.drop(['index'], axis=1)
  temp.append(df)
  df = walmar.filter(items=['Start_of_Week', 'CPI'])
  df = df.rename(columns={"Start_of_Week": "ds", "CPI": "y"})
  df = df.reset_index()
  df = df.drop(['index'], axis=1)
  cpi.append(df)
```

```python
zipbObj = zip((range(len(fuel))), fuel)
```

```python
fueldict = dict(zipbObj)
```

```python
fcst_all = pd.DataFrame() 
for name, df in fueldict.items():    
    if name > 0:
      m = Prophet().fit(df)
      future = m.make_future_dataframe(periods=55, freq= 'W')
      fcst = m.predict(future)
      fcst['Store'] = name 
      fcst_all = pd.concat((fcst_all, fcst))
```

```python
fuel_fcst_all = fcst_all
fuel_fcst_all
```

```python
zipbObj = zip((range(len(unem))), unem)
unemdict = dict(zipbObj)
```

```python
unem_fcst_all = pd.DataFrame() 
for name, df in unemdict.items():    
    if name > 0:
      m = Prophet().fit(df)
      future = m.make_future_dataframe(periods=55, freq = "W")
      fcst = m.predict(future)
      fcst['Store'] = name 
      unem_fcst_all = pd.concat((unem_fcst_all, fcst))
```

```python
zipbObj = zip((range(len(temp))), temp)
tempdict = dict(zipbObj)
```

```python
temp_fcst_all = pd.DataFrame() 
for name, df in tempdict.items():    
    if name > 0:
      m = Prophet().fit(df)
      future = m.make_future_dataframe(periods=55, freq = "W")
      fcst = m.predict(future)
      fcst['Store'] = name 
      temp_fcst_all = pd.concat((temp_fcst_all, fcst))
```

```python
zipbObj = zip((range(len(cpi))), cpi)
cpidict = dict(zipbObj)
```

```python
cpi_fcst_all = pd.DataFrame() 
for name, df in cpidict.items():    
    if name > 0:
      m = Prophet().fit(df)
      future = m.make_future_dataframe(periods=55, freq = "W")
      fcst = m.predict(future)
      fcst['Store'] = name 
      cpi_fcst_all = pd.concat((cpi_fcst_all, fcst))
```

```python
cpi_fcst_all
cpi_future = cpi_fcst_all.filter(items=['ds', 'yhat', "Store"])
```

```python
temp_future = temp_fcst_all.filter(items=['ds', 'yhat', "Store"])
unem_future = unem_fcst_all.filter(items=['ds', 'yhat', "Store"])
fuel_future = fuel_fcst_all.filter(items=['ds', 'yhat', "Store"])
```

```python
print(walmart['Start_of_Week'].iloc[[-1]].item())
print(cpi_future['ds'].tail(1).item())
```

```python
start_date = walmart['Start_of_Week'].iloc[[-1]].item()
end_date = cpi_future['ds'].tail(1).item()

after_start_date = cpi_future["ds"] > start_date
before_end_date = cpi_future["ds"] <= end_date
between_two_dates = after_start_date & before_end_date
cpi_predictions = cpi_future.loc[between_two_dates]

cpi_predictions
```

```python
start_date = walmart['Start_of_Week'].iloc[[-1]].item()
end_date = temp_future['ds'].tail(1).item()

after_start_date = temp_future["ds"] > start_date
before_end_date = temp_future["ds"] <= end_date
between_two_dates = after_start_date & before_end_date
temp_predictions = temp_future.loc[between_two_dates]

temp_predictions
```

```python
start_date = walmart['Start_of_Week'].iloc[[-1]].item()
end_date = unem_future['ds'].tail(1).item()

after_start_date = unem_future["ds"] > start_date
before_end_date = unem_future["ds"] <= end_date
between_two_dates = after_start_date & before_end_date
unem_predictions = unem_future.loc[between_two_dates]

unem_predictions
```

```python
start_date = walmart['Start_of_Week'].iloc[[-1]].item()
end_date = fuel_future['ds'].tail(1).item()

after_start_date = fuel_future["ds"] > start_date
before_end_date = fuel_future["ds"] <= end_date
between_two_dates = after_start_date & before_end_date
fuel_predictions = fuel_future.loc[between_two_dates]

fuel_predictions
```

```python
cpi_predictions = cpi_predictions.rename(columns={"yhat": "CPI", "ds": "Date"})
temp_predictions = temp_predictions.rename(columns={"yhat": "Temperature_C", "ds": "Date"})
unem_predictions = unem_predictions.rename(columns={"yhat": "Unemployment", "ds": "Date"})
fuel_predictions = fuel_predictions.rename(columns={"yhat": "Fuel_Price", "ds": "Date"})
```

```python
#temp_predictions = temp_predictions.reset_index().drop(['index'], axis=1)
cpi_predictions = cpi_predictions.reset_index().drop(['index'], axis=1)
fuel_predictions = fuel_predictions.reset_index().drop(['index'], axis=1)
unem_predictions = unem_predictions.reset_index().drop(['index'], axis=1)
```

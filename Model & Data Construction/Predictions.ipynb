```
import pandas as pd
import numpy as np
from fbprophet import Prophet
```

```
walmart = pd.read_csv('walmartdata.csv')
walmart
```

```
fuel = []
unem = []
temp = []
cpi = []
for i in range((walmart["Store"].max()) + 1):
  walmar = walmart.loc[walmart['Store'] == i]
  df = walmar.filter(items=['Start_of_Week', 'Fuel_Price'])
  df = df.rename(columns={"Start_of_Week": "ds", "Fuel_Price": "y"})
  df = df.reset_index()
  df = df.drop(['index'], axis=1)
  fuel.append(df)
  df = walmar.filter(items=['Start_of_Week', 'Unemployment'])
  df = df.rename(columns={"Start_of_Week": "ds", "Unemployment": "y"})
  df = df.reset_index()
  df = df.drop(['index'], axis=1)
  unem.append(df)
  df = walmar.filter(items=['Start_of_Week', 'Temperature_C'])
  df = df.rename(columns={"Start_of_Week": "ds", "Temperature_C": "y"})
  df = df.reset_index()
  df = df.drop(['index'], axis=1)
  temp.append(df)
  df = walmar.filter(items=['Start_of_Week', 'CPI'])
  df = df.rename(columns={"Start_of_Week": "ds", "CPI": "y"})
  df = df.reset_index()
  df = df.drop(['index'], axis=1)
  cpi.append(df)
```

```
zipbObj = zip((range(len(fuel))), fuel)
```

```
fueldict = dict(zipbObj)
```

```
fcst_all = pd.DataFrame() 
for name, df in fueldict.items():    
    if name > 0:
      m = Prophet().fit(df)
      future = m.make_future_dataframe(periods=55, freq= 'W')
      fcst = m.predict(future)
      fcst['Store'] = name 
      fcst_all = pd.concat((fcst_all, fcst))
```

```
fuel_fcst_all = fcst_all
fuel_fcst_all
```

```
zipbObj = zip((range(len(unem))), unem)
unemdict = dict(zipbObj)
```

```
unem_fcst_all = pd.DataFrame() 
for name, df in unemdict.items():    
    if name > 0:
      m = Prophet().fit(df)
      future = m.make_future_dataframe(periods=55, freq = "W")
      fcst = m.predict(future)
      fcst['Store'] = name 
      unem_fcst_all = pd.concat((unem_fcst_all, fcst))
```

```
zipbObj = zip((range(len(temp))), temp)
tempdict = dict(zipbObj)
```

```
temp_fcst_all = pd.DataFrame() 
for name, df in tempdict.items():    
    if name > 0:
      m = Prophet().fit(df)
      future = m.make_future_dataframe(periods=55, freq = "W")
      fcst = m.predict(future)
      fcst['Store'] = name 
      temp_fcst_all = pd.concat((temp_fcst_all, fcst))
```

```
zipbObj = zip((range(len(cpi))), cpi)
cpidict = dict(zipbObj)
```

```
cpi_fcst_all = pd.DataFrame() 
for name, df in cpidict.items():    
    if name > 0:
      m = Prophet().fit(df)
      future = m.make_future_dataframe(periods=55, freq = "W")
      fcst = m.predict(future)
      fcst['Store'] = name 
      cpi_fcst_all = pd.concat((cpi_fcst_all, fcst))
```
